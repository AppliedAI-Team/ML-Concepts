Let us consider Data Matrix X is defines as below:
$$X_{(N,d)} = 
 \begin{pmatrix}
  x_{11} & x_{12} & \cdots & x_{1d} \\
  x_{21} & x_{22} & \cdots & x_{2d} \\
  \vdots  & \vdots  & \ddots & \vdots  \\
  x_{N1} & x_{N2} & \cdots & x_{Nd} 
 \end{pmatrix}_{(N,d)}$$
 
Where 
* d = Dimensionality of the dataset
* N = Number of data points in the dataset

$$X_{(N,d)} = 
 \begin{pmatrix}
  x^T_1\\
  x^T_2\\
  \vdots\\
  x^T_N
 \end{pmatrix}_{(N,d)}$$

Where
* $x^T_1 = \text {d dimensional vector/data point/observation} = <x_{11} , x_{12} , \cdots , x_{1d}>$

To Reduce the dimensionality of the dataset from $d$ dimensions to the $d'$:
1. Calculate the Covariance matrix of $X_{(N,d)}$ ===> $S_{(d,d)}$
2. Calculate the d number of Eigen Values and Eigen Vectors.
3. Draw the Explained variance vs NUmber of Eigen vectors graph to answer a question below
How many number of Eigen Vectors should i consider to preserve x% of information in the given dataset while reducing the dimensionalty from d to d'(Eigen values play an important role in this step)
4. Transform the datset based on the Eigen vectors.

## Calculate the Covariance matrix of $X_{(N,d)}$ ===> $S_{(d,d)}$

We know that $S_{(d,d)} = \frac {1} {n-1} X^T_{(d,N)}X_{(N,d)}$

And via computation we can get Eigen values($\lambda _1, \lambda _2, \cdots , \lambda _d$), Eigen vectors($v_1, v_2, \cdots, v_d$)

Where $\lambda _i$ is a scalar and the $v_i$ is a column vector of $(d,1)$ dimensionality.

## Draw the Explained variance vs Number of Eigen vectors graph

If dimnesionality of the datast d = 20 then the graph will be as shown below.



As you can see that by using just 10 features we are able to capture 99% of information in the datset. Now it is your call as a Machine Learning Practitioner to make a choice.

1. If you want to preserve 99% of information then you can make d' equal to 10.
2. If you want to preserve 90% of information then you can make d' equal to 6.

There is nothing wrong or right in this procedure, you should choose the threshold by using the Domain Knowledge or by making it as a hyperparameter and choice will depends upon th performance of the ML model trained on features created by making d' as a choice of threshold on the CV Dataset.
$\color{red}{\textbf{Let us consider d' = 6 (by our choice)}}$

## Transform the datset based on the Eigen vectors  $X_{(N,d)} ===> X'_{(N,d')}$

Here d'= 6 

Number of Eigen Vcetors to consider = 6

Eigen vector Matrix => $U_{(d,6)}$

$ U_{(d,6)} = 
 \begin{pmatrix}
  \Uparrow & \Uparrow & \Uparrow & \Uparrow \\
  v_1 & v_2 & \cdots & v_6 \\
  \Downarrow  & \Downarrow  & \Downarrow & \Downarrow  \\ 
 \end{pmatrix}_{(d,6)} $
 
 Where Each vector $v_i$ is d dimensional column vector $$v_{i} = 
 \begin{pmatrix}
  v_{i1}\\
  v_{i2}\\
  \vdots\\
  v_{i6}
 \end{pmatrix}_{(6,1)}$$
 
$ X'_{(N,6)} =  X_{(N,d)}U_{(d,6)}$

$ X'_{(N,6)} =  \begin{pmatrix}
  x_{11} & x_{12} & \cdots & x_{1d} \\
  x_{21} & x_{22} & \cdots & x_{2d} \\
  \vdots  & \vdots  & \ddots & \vdots  \\
  x_{N1} & x_{N2} & \cdots & x_{Nd} 
 \end{pmatrix}_{(N,d)}\begin{pmatrix}
  \Uparrow & \Uparrow & \Uparrow & \Uparrow \\
  v_1 & v_2 & \cdots & v_6 \\
  \Downarrow  & \Downarrow  & \Downarrow & \Downarrow  \\ 
 \end{pmatrix}_{(d,6)}$
 
$\color{red}{\textbf{If you want to Transform a single datapoint x_q of (d,1) dimensionality to (6,1) at production}}$

$ x_{\text {q dash}}^T =  x_q^TU_{(d,6)}$

$ x_q^T =  \begin{pmatrix}
